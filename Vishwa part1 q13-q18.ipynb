```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use('default')
```

```python
brent = pd.read_csv('Brent Oil.csv')
gold = pd.read_csv('Gold 100years.csv')
silver = pd.read_csv('silver 100 years.csv')
```

```python
print(brent.head())
print(gold.head())
print(silver.head())

print(brent.columns)
print(gold.columns)
print(silver.columns)
```

```python
brent['Date'] = pd.to_datetime(brent['Date'], dayfirst=True)
gold['Date'] = pd.to_datetime(gold['Date'], dayfirst=True)
silver['Date'] = pd.to_datetime(silver['Date'], dayfirst=True)
```

```python
# =========================================================
# Question 13
# Create a line chart for Brent Oil prices over time
# =========================================================

plt.figure(figsize=(12,6))

plt.plot(
    brent['Date'],
    brent['Value'],
    color='darkblue',
    linewidth=2
)

plt.title(
    'Brent Oil Prices Over Time',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Year', fontsize=14)
plt.ylabel('Price', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
print("Interpretation")
print("""
1. 1946 — Lowest Oil Prices
After World War II, most countries were still rebuilding their economies and industries, so people were using much less oil.
Since demand was low and supply was enough, oil prices stayed very cheap.

Price in 1946: 21.11


2. 1970 to 1980 — Sharp Increase in Oil Prices
Oil prices increased a lot because OPEC countries reduced oil exports during political conflicts in the Middle East.
The Iranian Revolution also affected oil production, which made countries fear shortages and caused prices to rise quickly.

Price in 1970: 29.26
Price in 1980: 141.57


3. 1982 to 1990 — Decline in Oil Prices
After the oil crises, many countries started saving fuel and using energy more carefully.
At the same time, more oil was produced in places outside OPEC, so prices slowly started falling.

Price in 1982: 118.53
Price in 1990: 70.29


4. 1990 — Increase in Oil Prices
Oil prices went up again during the Gulf War when Iraq invaded Kuwait.
Since both countries were major oil producers, people worried that oil supply could be affected, which increased prices.

Price at start of 1990: 58.81
Highest price in 1990: 98.37


5. 1990s to Early 2000s — Fluctuations and Recovery
During the late 1990s, oil prices moved up and down because some countries faced economic problems and used less fuel.
In the early 2000s, fast growth in countries like China and India increased the need for oil again.

Price in 1991: 46.77
Price in 2004: 115.39


6. 2008 — Extreme Increase in Oil Prices
In 2008, oil prices reached very high levels because countries like China and India were growing rapidly and using huge amounts of fuel.
At the same time, supply could not increase fast enough, and investors expected prices to rise even more.

Price at start of 2008: 143.41
Highest price in 2008: 211.21


7. Late 2008 — Sharp Fall in Oil Prices
Later in 2008, the global financial crisis slowed businesses, industries, and transportation around the world.
Since people and companies started using less fuel, oil prices dropped very quickly.

Highest price in 2008: 211.21
Price at end of 2008: 70.05


8. 2020 — Major Decline During COVID-19
During COVID-19, lockdowns stopped flights, travel, tourism, and many factory activities across the world.
Because fewer people were travelling and industries slowed down, oil demand fell sharply and prices decreased.

Price at start of 2020: 66.02
Lowest price in 2020: 24.77


9. 2021 to 2026 — Recovery, War, and Increase in Oil Prices
After COVID-19, industries, transportation, and flights restarted, which increased global oil demand very quickly.
The Russia–Ukraine war in 2022 created fears of oil supply shortages because Russia is one of the world’s largest oil producers.
Since demand increased faster than supply, Brent oil prices rose sharply during this period.

Price in 2021: 65.85
Latest price in dataset: 102.86
""")

```

```python
# =========================================================
# Question 14
# Create a line chart for Gold prices over time
# =========================================================

plt.figure(figsize=(12,6))

plt.plot(
    gold['Date'],
    gold['Value'],
    color='gold',
    linewidth=2
)

plt.title(
    'Gold Prices Over Time',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Year', fontsize=14)
plt.ylabel('Gold Price', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()
print("""
1. 1915 to 1930 — Stable Gold Prices
Gold prices remained stable because most countries followed the Gold Standard system, where governments controlled currency values using gold reserves.
Since prices were fixed by governments and markets were stable, gold prices did not fluctuate much.


2. 1930s — Great Depression and Increase in Gold Demand
During the Great Depression, banks failed and unemployment increased, making people lose trust in financial systems.
Investors started buying gold as a safer investment, which increased gold prices.


3. 1971 — Bretton Woods System Collapse
In 1971, the United States ended the Bretton Woods system, removing the direct link between the US dollar and gold.
After this, gold prices started moving freely based on market demand and inflation fears.


4. 1973 to 1980 — Oil Crisis, Inflation, and Political Tensions
The 1973 Oil Crisis and the Iranian Revolution increased oil prices and global inflation sharply.
Because investors feared economic instability, many people bought gold for safety, causing prices to rise rapidly.


5. 1980 — Historic Gold Peak
Gold prices reached a major peak around 1980 because inflation, oil shortages, and geopolitical tensions created panic in global markets.
Investors heavily purchased gold since they believed it was safer than currencies and stocks during uncertain times.



6. 1980 to 2000 — Long Stability and Decline
After 1980, economies became more stable and stock markets performed strongly, reducing investor interest in gold.
Higher interest rates also made bank savings and bonds more attractive than gold investments.



7. 2000 to 2008 — Recovery and Rising Demand
Gold prices started increasing again because of global uncertainty, wars, and a weakening US dollar.
Rapid economic growth in countries like China and India also increased gold demand for jewelry and investments.



8. 2008 — Global Financial Crisis
During the 2008 financial crisis, banks collapsed and stock markets crashed, making investors lose confidence in financial systems.
Many people moved their money into gold because it was considered a safer investment during the crisis.



9. 2011 — Another Major Gold Peak
Gold prices reached another peak around 2011 because countries in Europe were facing major debt problems and economic uncertainty.
Investors feared inflation and unstable financial markets, which increased gold demand further.



10. 2020 — COVID-19 Pandemic
During COVID-19, lockdowns stopped businesses, travel, and industries across the world, creating fears of a global recession.
Governments printed large amounts of money to support economies, which increased inflation concerns and pushed gold prices higher.



11. 2021 to 2026 — Inflation, Wars, and Rising Gold Prices
After COVID-19, inflation increased because demand recovered faster than supply, while wars and geopolitical tensions created uncertainty in global markets.
In India, rising gold imports and a weaker rupee made gold more expensive, which is why PM Modi recently requested people to reduce unnecessary gold purchases temporarily.


""")
```

```python
# =========================================================
# Question 15
# Create a line chart for Silver prices over time
# =========================================================

plt.figure(figsize=(12,6))

plt.plot(
    silver['Date'],
    silver['Value'],
    color='gray',
    linewidth=2
)

plt.title(
    'Silver Prices Over Time',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Year', fontsize=14)
plt.ylabel('Silver Price', fontsize=14)

plt.grid(True, linestyle='--', alpha=0.5)

plt.show()

print("""
1. 1915 to 1930 — Stable Silver Prices
Silver prices remained stable because many countries still used silver and gold in their monetary systems, and governments controlled financial markets strongly.
Industrial use of silver was also limited during this period, so demand and supply remained balanced.



2. 1930s — Great Depression and Weak Industrial Demand
During the Great Depression, factories slowed down, unemployment increased, and global economic activity reduced heavily.
Since silver is widely used in industries and manufacturing, reduced industrial demand kept silver prices low.



3. 1960s to Early 1970s — Industrial Growth Increased Silver Demand
Silver prices slowly increased because industries started using more silver in electronics, photography, and electrical equipment.
At the same time, inflation increased slightly, which also increased investor interest in precious metals.



4. 1973 to 1980 — Oil Crisis, Inflation, and Hunt Brothers Silver Bubble
The 1973 Oil Crisis increased inflation worldwide, making investors buy silver and gold to protect their money.
At the same time, the Hunt Brothers purchased massive amounts of silver, creating panic buying and pushing silver prices sharply upward.



5. 1980 — Silver Thursday Crash
After silver prices reached record highs, governments introduced stricter trading rules to stop market manipulation.
The Hunt Brothers were forced to sell silver quickly, causing panic selling and a major market crash called “Silver Thursday.”



6. 1980 to 2000 — Long Stability and Lower Prices
After the 1980 crash, investors lost confidence in silver markets, while stock markets and economies became stronger.
Since investors preferred stocks and bonds over silver, prices remained low and stable for many years.



7. 2000 to 2008 — Industrial Growth and Rising Silver Demand
Silver prices started increasing again because industries used more silver in electronics, batteries, medical equipment, and solar panels.
Rapid growth in countries like China and India also increased global silver demand.


8. 2008 — Global Financial Crisis
During the financial crisis, industries slowed down and silver demand initially decreased.
Later, investors feared inflation and financial instability, so they started buying silver again as a safer investment.



9. 2011 — Major Silver Rally
Silver prices increased sharply because investors feared inflation after governments printed large amounts of money during the financial crisis.
Many investors also saw silver as a cheaper alternative to gold, increasing demand further.


10. 2020 — COVID-19 Pandemic
During COVID-19 lockdowns, factories and industries stopped operating, which initially reduced silver demand.
Later, inflation fears and economic uncertainty increased investor demand for silver again.



11. 2021 to 2026 — Technology Demand, Inflation, and Rising Prices
After COVID-19, silver demand increased strongly because it is widely used in solar panels, electric vehicles, electronics, and renewable energy industries.
Inflation, wars, and a weaker Indian rupee also increased silver prices and made imports more expensive in India.


""")



```

```python
# =========================================================
# question 16: Create one combined chart showing Gold and Silver trends. 
# =========================================================

fig, ax1 = plt.subplots(figsize=(12,6))

# Gold axis
ax1.plot(
    gold['Date'],
    gold['Value'],
    color='gold',
    linewidth=2,
    label='Gold'
)

ax1.set_xlabel('Year', fontsize=14)
ax1.set_ylabel('Gold Price', color='gold', fontsize=14)

# Silver axis
ax2 = ax1.twinx()

ax2.plot(
    silver['Date'],
    silver['Value'],
    color='gray',
    linewidth=2,
    label='Silver'
)

ax2.set_ylabel('Silver Price', color='gray', fontsize=14)

plt.title(
    'Gold vs Silver Price Trends',
    fontsize=18,
    fontweight='bold'
)

ax1.grid(True, linestyle='--', alpha=0.5)

plt.show()

print("""
Gold vs Silver — Similarities

1. Both Increased Strongly After COVID-19
After the COVID-19 pandemic, prices of both gold and silver increased because inflation rose and economies slowly recovered.
Investors purchased more precious metals due to uncertainty, rising oil prices, and geopolitical tensions.

2. Both Are Affected by Global Events
Major events such as wars, oil crises, financial crises, and pandemics affected both gold and silver prices.
During uncertain situations, people invest more in precious metals, which increases demand and prices.


Gold vs Silver — Differences

1. Gold Is Mainly an Investment Asset While Silver Is Also an Industrial Metal
Gold is mainly used for investment, jewelry, and central bank reserves, so its price mostly depends on investor confidence and economic conditions.
Silver is widely used in industries like electronics, solar panels, batteries, and electric vehicles, so industrial demand strongly affects silver prices.

2. Silver Prices Are More Volatile Than Gold Prices
Gold prices usually move more steadily because gold markets are larger and considered more stable by investors.
Silver prices fluctuate more because the silver market is smaller and more sensitive to industrial demand, speculation, and economic changes.
""")


```

```python
# =========================================================
# Question 17
# Create a bar chart showing average annual Brent Oil prices
# for the most recent 10 years
# =========================================================

brent['Year'] = brent['Date'].dt.year

annual_avg = brent.groupby('Year')['Value'].mean().reset_index()

recent_10 = annual_avg.tail(10)

plt.figure(figsize=(12,6))

plt.bar(
    recent_10['Year'].astype(str),
    recent_10['Value'],
    color='teal'
)

plt.title(
    'Average Annual Brent Oil Prices (Recent 10 Years)',
    fontsize=18,
    fontweight='bold'
)

plt.xlabel('Year', fontsize=14)
plt.ylabel('Average Price', fontsize=14)

plt.xticks(rotation=45)

plt.grid(axis='y', linestyle='--', alpha=0.5)

plt.show()
print("""
1. 2017 — Increase in Oil Prices
Oil prices increased because global demand improved as many economies recovered and industries expanded.
OPEC and Russia also reduced oil production to control supply and support prices.

2. 2018 — Further Increase in Oil Prices
Prices increased further due to strong global economic growth and higher fuel demand.
Political tensions involving Iran and supply concerns in the Middle East also pushed prices upward.

3. 2019 — Slight Decrease in Oil Prices
Oil prices slightly decreased because of slowing global economic growth and trade tensions between the United States and China.
Investors worried that lower industrial activity would reduce oil demand.

4. 2020 — Sharp Fall During COVID-19
Oil prices dropped sharply because the COVID-19 pandemic stopped travel, tourism, flights, and many industries worldwide.
Since people stayed home during lockdowns, global oil demand collapsed.

5. 2021 — Recovery in Oil Prices
Oil prices recovered as countries reopened after the pandemic and transportation activities increased again.
Demand rose faster than oil supply, which pushed prices upward.

6. 2022 — Major Increase in Oil Prices
Prices increased strongly because of the Russia–Ukraine war, which created major supply concerns in global markets.
Many countries also faced inflation and energy shortages during this period.

7. 2023 — Stabilization of Oil Prices
Oil prices became more stable as supply improved and economies adjusted after earlier disruptions.
However, OPEC production cuts and geopolitical tensions still kept prices relatively high.

8. 2024 — Slight Decline in Oil Prices
Prices slightly decreased because some countries increased oil production and demand growth slowed down.
Economic uncertainty in different regions also reduced fuel consumption growth.

9. 2025 — Further Decline in Oil Prices
Oil prices declined further due to improved supply conditions and slower global economic activity.
Many countries also focused more on renewable energy and fuel-efficient technologies.

10. 2026 — Increase in Oil Prices Again
Prices increased again because global demand recovered while supply growth remained limited.
Geopolitical tensions and production controls by major oil-producing countries also supported higher prices.
""")

```

```python
# =========================================================
# Question 18
# Create a simple table showing the latest available value
# for Brent Oil, Gold and Silver
# =========================================================

latest_brent = brent.iloc[-1]
latest_gold = gold.iloc[-1]
latest_silver = silver.iloc[-1]

latest_table = pd.DataFrame({
    'Commodity': ['Brent Oil', 'Gold', 'Silver'],
    'Latest Date': [
        latest_brent['Date'],
        latest_gold['Date'],
        latest_silver['Date']
    ],
    'Latest Value': [
        latest_brent['Value'],
        latest_gold['Value'],
        latest_silver['Value']
    ]
})

print("\nLatest Available Values Table:\n")

display(latest_table.style
        .set_properties(**{
            'background-color': '#f5f5f5',
            'color': 'black',
            'border-color': 'black',
            'font-size': '12pt'
        })
        .set_table_styles([
            {
                'selector': 'th',
                'props': [
                    ('background-color', '#003366'),
                    ('color', 'white'),
                    ('font-size', '13pt')
                ]
            }
        ]))
```
